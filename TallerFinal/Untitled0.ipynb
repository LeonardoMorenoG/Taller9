```python
def likelihood(y2_obs, y_model, x_obs2, x_model,):
    chi_squared = sum((y_obs-y_model)**2)+sum((x_obs-x_model)**2)
    return (-chi_squared)
```

```python
def my_modely(y1, z1, y, z, z2):
    return ((y1 - y)/(z1- z))*z2 + y1
```

```python
def my_modelx(x1, z1, x, z,z2):
    return ((x1 - x)/(z1- z))* z2 + x1
```

```python
x_walk = empty((0))
y_walk = empty((0))
z_walk = empty((0))

x_walk = append(x_walk, random.random())
y_walk = append(y_walk, random.random())
z_walk = -1*append(z_walk, random.random())

print x_walk
print y_walk
print z_walk
```

```python
n_iterations = 200 #this is the number of iterations I want to make

for i in range(n_iterations):
        
    x_prime = np.random.normal(x_walk[i], 0.1)
    y_prime = np.random.normal(y_walk[i], 0.1) 
    z_prime = np.random.normal(z_walk[i], 0.1)

    y_init = my_modely(y_obs,z_obs ,y_walk, z_walk, z2_obs)
    x_init = my_modelx(x_obs,z_obs ,x_walk, z_walk, z2_obs)
    y_prime = my_modely(y_obs,z_obs ,y_prime, z_prime, z2_obs)
    x_prime = my_modelx(x_obs,z_obs ,x_prime, z_prime, z2_obs)
    
    alpha = likelihood(y2_obs, y_prime, x2_obs, x_prime )/likelihood(y2_obs, y_init, x2_obs, x_init)
    if(alpha<=1.0):
        x_walk  = append(x_walk,x_prime)
        y_walk  = append(y_walk,y_prime)
        z_walk  = append(z_walk,z_prime)
    else:
        beta = random.random()
        if(beta<=exp(-alpha)):
            x_walk  = append(x_walk,x_prime)
            y_walk  = append(y_walk,y_prime)
            z_walk  = append(z_walk,z_prime)
        else:
            x_walk  = append(x_walk,x_walk[i])
            y_walk  = append(y_walk,y_walk[i])
            z_walk  = append(z_walk,z_walk[i])
```
